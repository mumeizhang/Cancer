4.1  Method I using ggplot

```r
mytheme<- theme_bw() + theme(plot.title=element_text(hjust=0.5),axis.title=element_text(size=rel(1)),axis.text=element_text(size=rel(1)),legend.position = "top")
xx<-ggplot(data=cancer2,mapping=aes(x=Group,y=Secondary.Metastasis.Rate, color=Day,fill=Day))
xx+geom_boxplot(position="dodge")+labs(x="Group",y="Second-metatstasis rate")+
  theme_bw()+scale_fill_brewer(palette="Pastel1")+geom_jitter(size=0.5,alpha=0.5)+coord_fixed(ratio=0.02)
```

4.2 Method II using ggboxplot

```r
ggboxplot(data=cancer2,x="Group",y="Secondary.Metastasis.Rate",color="Day",fill = "Day",palette = "ncp", add = "jitter",shape="Group",
          scale="free", ylab="Second-metatstasis rate")
ggboxplot(data=cancer2,x="Group",y="Secondary.Metastasis.Rate",color="Group",fill = "Group",palette = "ncp", add = "jitter",shape="Group",
          facet.by="Day", scale="free",ylab="Second-metatstasis rate")+stat_compare_means()
```

### Conlusion: At 0.05 significant level, there is signicant difference between the treatment group and placebo group
